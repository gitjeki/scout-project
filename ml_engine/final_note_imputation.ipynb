```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

data = {
    'Age': [25, 45, 30, 50],
    'Education': [1, 3, 2, 3],
    'Income': [40000, 70000, np.nan, 85000]
}
df = pd.DataFrame(data)

print("### Original Data with Missing Value (Customer C) ###")
print(df)

imputer = KNNImputer(n_neighbors=2)

df_imputed_array = imputer.fit_transform(df)

df_knn_imputed = pd.DataFrame(df_imputed_array, columns=df.columns)

print("\n---")
print(f"### KNN Imputed Data (K=2, Income for Customer C) ###")
print(df_knn_imputed)
```

```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

data = {
    'Age': [25, 45, 30, 50],
    'Education': [1, 3, 2, np.nan],
    'Income': [40000, 70000, np.nan, 85000]
}
df = pd.DataFrame(data)

print("\n---")
print("### Original Data with Mixed Missing Values (C & D) ###")
print(df)

rf_regressor = RandomForestRegressor(random_state=42, n_estimators=100, max_depth=5)
imputer_reg = IterativeImputer(estimator=rf_regressor, random_state=42, max_iter=10)
df_imputed_array = imputer_reg.fit_transform(df)

df_missforest_imputed = pd.DataFrame(df_imputed_array, columns=df.columns).round(0) # Rounding for display

print("\n---")
print(f"### MissForest-like Imputed Data (with RF Regressor) ###")
print(df_missforest_imputed)
```
